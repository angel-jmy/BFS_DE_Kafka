```python
import math
import pandas as pd

path = 'Employee_Salaries.csv'
depts = {"ECC", "CIT", "EMS"}
```

```python
df = pd.read_csv(path)
df.head()
```

```python

sub_df = df[df.Department.isin(depts)].copy()
sub_df['Initial Hire Date'] = pd.to_datetime(sub_df['Initial Hire Date'], format = '%d-%b-%Y')
sub_df.Salary = sub_df.Salary.astype(int)
sub_df = sub_df[sub_df['Initial Hire Date'].dt.year >= 2010].copy()
sub_df.shape
```

```python
sub_df.groupby('Department').agg(
    tot_salary = ('Salary', 'sum')
)
```
